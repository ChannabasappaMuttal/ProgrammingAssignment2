### Build the `text-to-speech` pipeline using 🤗 Transformers Library

````python
from transformers.utils import logging

logging.set_verbosity_error()
````

````python
from transformers import pipeline

narrator = pipeline("text-to-speech",
                    model="kakao-enterprise/vits-ljs")
````

_Note_: Find more information about the model 'kakao-enterprise/vits-ljs', [here](https://huggingface.co/kakao-enterprise/vits-ljs).

````python
# Set a text sample
text = """KLE Technological University (KLE Tech) has its roots in one of the premier Engineering Institutions of Karnataka,
B. V. Bhoomaraddi College of Engineering and Technology (BVB), a prestigious Engineering College in Hubballi.
The founding organization KLE Society, Belgaum, established BVB College in 1947 to create an institution that would lay the foundation of modern
engineering education in the northern region of Karnataka."""
````

````python
# Call the text-to-speech pipeline and store the result in narrated_text
narrated_text = narrator(text) # This line is added to call the pipeline and store the result
````

````python
# Listen to the text converted to audio
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
````
